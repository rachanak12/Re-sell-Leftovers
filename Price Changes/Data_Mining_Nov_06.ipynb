# Data Mining

In this project, I used the Google Trends API to retrieve data for the past 7 days, including November 6. This allowed me to analyze search trends and extract valuable insights into user behavior and interests over this specific time period.

```python
keywords = [
    "pizza", "burger", "sushi", "pasta", "salad", "steak", "tacos", "sandwich", "noodles", "omelette",
    "burrito", "risotto", "dumplings", "ravioli", "lasagna", "paella", "curry", "falafel", "quiche", "wrap",
    "gyros", "shawarma", "bruschetta", "calzone", "fajitas", "enchiladas", "gnocchi", "meatballs", "wings", "pancakes",
    "waffles", "crepes", "donuts", "tiramisu", "cheesecake", "brownie", "bread", "apple", "banana", "orange", "grape", "strawberry", "blueberry", "pineapple", "kiwi", "mango", "pear",
        "bread", "milk", "eggs", "butter", "cheese",
    "yogurt", "chicken", "beef", "pork", "fish",
    "rice", "pasta", "vegetables", "fruits", "cereal",
    "sugar", "salt", "pepper", "olive oil", "vinegar"
]
```

```python
import pandas as pd
all_data = pd.DataFrame()
```

```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='en-US', tz=360)
for i in range(0, len(keywords), 4):
    keyword_group = keywords[i:i+4]
    pytrends.build_payload(keyword_group, timeframe='now 7-d')
    df = pytrends.interest_over_time()
    all_data = pd.concat([all_data, df], axis=1)
print(all_data)
```

```python
all_data = all_data.loc[:, ~all_data.columns.str.startswith('isPartial')]
```

```python
all_data['total'] = all_data.sum(axis=1)
```

```python
all_data.reset_index(drop=False, inplace=True)
```

```python
all_data
```

```python
import matplotlib.pyplot as plt
```

```python
all_data['hour'] = all_data['date'].dt.hour

plt.figure(figsize=(8, 5))
plt.plot(all_data['hour'], all_data['total'], marker='o', color='blue', label='Total')
plt.title('Total Value Distribution by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Total')
plt.axhline(y=all_data['total'].mean(), color='red', linestyle='--', label='Mean Total')
plt.legend()
plt.grid()
plt.xticks(all_data['hour'])
plt.show()
```

```python
all_data["date"]=pd.to_datetime(all_data["date"])
all_data['is_holiday'] = all_data['date'].dt.strftime('%m-%d').isin(['11-02', '11-03']).astype(int)
```

```python
all_data.to_csv('all_data_nov_06.csv', index=False)
```
